This notebook performs comprehensive data quality checks on the Customer Demographics dataset.

## 1. Data Loading

```python
import pandas as pd

# Load the Customer Demographics data
df_customer = pd.read_csv('../datasets/Customer_Demographics.csv')
print("Data loaded successfully!")
print(f"Total records: {len(df_customer)}")
```

## 3. Data Quality Checks

### 3.1 Missing Values Analysis

```python
# Check for missing values
missing_values = df_customer.isnull().sum()
missing_percentage = (df_customer.isnull().sum() / len(df_customer)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})
print("=== Missing Values Analysis ===")
print(missing_df)
print(f"\nTotal missing values: {df_customer.isnull().sum().sum()}")
```

### 3.2 Duplicate Records Check

```python
# Check for duplicate records
duplicates_all = df_customer.duplicated().sum()
print("=== Duplicate Records Check ===")
print(f"Total duplicate rows: {duplicates_all}")

# Check for duplicate Customer IDs
duplicates_customer_id = df_customer['Customer ID'].duplicated().sum()
print(f"Duplicate Customer IDs: {duplicates_customer_id}")

if duplicates_customer_id > 0:
    print("\nDuplicate Customer IDs found:")
    print(df_customer[df_customer['Customer ID'].duplicated(keep=False)].sort_values('Customer ID'))
```

### 3.3 Data Type Validation

```python
# Check data types
print("=== Data Type Validation ===")
print(df_customer.dtypes)

# Verify expected data types
expected_types = {
    'Customer ID': 'object',
    'Age': 'int64',
    'Gender': 'object',
    'Income Bracket': 'object',
    'Purchase Frequency': 'int64',
    'Average Spend': 'float64'
}

type_issues = []
for col, expected in expected_types.items():
    if df_customer[col].dtype != expected:
        type_issues.append(f"{col}: expected {expected}, got {df_customer[col].dtype}")

if type_issues:
    print("\nData Type Issues:")
    for issue in type_issues:
        print(f"  - {issue}")
else:
    print("\nAll data types are correct!")
```

### 3.4 Value Range Checks

```python
# Check value ranges for numeric columns
print("=== Value Range Checks ===")

# Age range check (typical adult age: 18-100)
age_min, age_max = df_customer['Age'].min(), df_customer['Age'].max()
print(f"\nAge: Min={age_min}, Max={age_max}")
age_issues = df_customer[(df_customer['Age'] < 18) | (df_customer['Age'] > 100)]
print(f"Age values outside 18-100 range: {len(age_issues)}")

# Purchase Frequency range check (typically 1-12 months)
freq_min, freq_max = df_customer['Purchase Frequency'].min(), df_customer['Purchase Frequency'].max()
print(f"\nPurchase Frequency: Min={freq_min}, Max={freq_max}")
freq_issues = df_customer[(df_customer['Purchase Frequency'] < 1) | (df_customer['Purchase Frequency'] > 12)]
print(f"Purchase Frequency outside 1-12 range: {len(freq_issues)}")

# Average Spend check (should be positive)
spend_min, spend_max = df_customer['Average Spend'].min(), df_customer['Average Spend'].max()
print(f"\nAverage Spend: Min={spend_min:.2f}, Max={spend_max:.2f}")
spend_issues = df_customer[df_customer['Average Spend'] <= 0]
print(f"Non-positive Average Spend values: {len(spend_issues)}")
```

### 3.5 Categorical Value Validation

```python
# Check unique values in categorical columns
print("=== Categorical Value Validation ===")

# Gender validation
valid_genders = ['Male', 'Female', 'Other']
unique_genders = df_customer['Gender'].unique()
print(f"\nGender unique values: {unique_genders}")
invalid_genders = df_customer[~df_customer['Gender'].isin(valid_genders)]
print(f"Invalid Gender values: {len(invalid_genders)}")

# Income Bracket validation
valid_income_brackets = ['1-5 LPA', '10 LPA', '10-20 LPA', '20+ LPA']
unique_income = df_customer['Income Bracket'].unique()
print(f"\nIncome Bracket unique values: {unique_income}")
invalid_income = df_customer[~df_customer['Income Bracket'].isin(valid_income_brackets)]
print(f"Invalid Income Bracket values: {len(invalid_income)}")
```

### 3.6 Outlier Detection

```python
# Detect outliers using IQR method
print("=== Outlier Detection (IQR Method) ===")

def detect_outliers_iqr(column):
    Q1 = df_customer[column].quantile(0.25)
    Q3 = df_customer[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_customer[(df_customer[column] < lower_bound) | (df_customer[column] > upper_bound)]
    return len(outliers), lower_bound, upper_bound

# Check outliers for numeric columns
numeric_columns = ['Age', 'Purchase Frequency', 'Average Spend']
for col in numeric_columns:
    count, lower, upper = detect_outliers_iqr(col)
    print(f"\n{col}:")
    print(f"  Outliers: {count}")
    print(f"  Lower Bound: {lower:.2f}")
    print(f"  Upper Bound: {upper:.2f}")
```

## 4. Data Quality Summary

```python
# Generate comprehensive data quality summary
print("=" * 50)
print("DATA QUALITY SUMMARY")
print("=" * 50)

print(f"\n1. Dataset Overview:")
print(f"   - Total Records: {len(df_customer)}")
print(f"   - Total Columns: {len(df_customer.columns)}")

print(f"\n2. Missing Values:")
print(f"   - Total Missing: {df_customer.isnull().sum().sum()}")
for col in df_customer.columns:
    missing = df_customer[col].isnull().sum()
    print(f"   - {col}: {missing}")

print(f"\n3. Duplicate Records:")
print(f"   - Duplicate Rows: {df_customer.duplicated().sum()}")
print(f"   - Duplicate Customer IDs: {df_customer['Customer ID'].duplicated().sum()}")

print(f"\n4. Data Quality Status:")
quality_passed = True
if df_customer.isnull().sum().sum() > 0:
    print(f"   ❌ Missing values found")
    quality_passed = False
else:
    print(f"   ✓ No missing values")

if df_customer.duplicated().sum() > 0:
    print(f"   ❌ Duplicate records found")
    quality_passed = False
else:
    print(f"   ✓ No duplicate records")

if len(age_issues) > 0 or len(freq_issues) > 0 or len(spend_issues) > 0:
    print(f"   ❌ Value range issues found")
    quality_passed = False
else:
    print(f"   ✓ All values within expected ranges")

if len(invalid_genders) > 0 or len(invalid_income) > 0:
    print(f"   ❌ Invalid categorical values found")
    quality_passed = False
else:
    print(f"   ✓ All categorical values are valid")

if quality_passed:
    print(f"\n✅ DATA QUALITY CHECK PASSED!")
else:
    print(f"\n❌ DATA QUALITY ISSUES DETECTED - REVIEW REQUIRED!")
```
